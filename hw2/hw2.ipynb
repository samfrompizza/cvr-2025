## Подсчёт числа входящих людей

```python
import numpy as np
import cv2

PERIOD = 70

video_path = "Door.mp4"
cam = cv2.VideoCapture(video_path)
_, background = cam.read()
background = cv2.cvtColor(background, cv2.COLOR_BGR2HSV)

persons_detected = 0
detection_time = 0
cooldown = 8
frame_rate = cam.get(cv2.CAP_PROP_FPS)
frames_cooldown = int(frame_rate * cooldown)

while(True):
    success, frame = cam.read()

    if (not success):
        break

    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    #avg = 0.96 * avg + 0.04 * frame
    background = cv2.addWeighted(background, 0.96, frame, 0.04, 0)
    diff = cv2.absdiff(hsv, background)
    foreground_mask = cv2.inRange(diff, (20, 20, 20), (255, 255, 255))

    # Erosion
    kernel = np.ones((9, 9), np.uint8)
    morph = cv2.morphologyEx(foreground_mask, cv2.MORPH_CLOSE, kernel)
    
    # Analysing connected components
    connectivity = 5
    output = cv2.connectedComponentsWithStats(morph, connectivity, cv2.CV_32S)
    num_components = output[0]
    components = output[1]
    stats = output[2]

    filtered = np.zeros_like(morph)
    detected = False

    for i in range(num_components):
        top    = stats[i, cv2.CC_STAT_TOP]
        left   = stats[i, cv2.CC_STAT_LEFT]
        width  = stats[i, cv2.CC_STAT_WIDTH]
        height = stats[i, cv2.CC_STAT_HEIGHT]
        area   = stats[i, cv2.CC_STAT_AREA]

        if area >= 20000:
            detected = True
            filtered[np.where(components == i)] = 255

            cv2.rectangle(frame, (left, top), (left + width, top + height), (0, 255, 0), 2)
        
    current_time = cam.get(cv2.CAP_PROP_POS_FRAMES)
    if detected and (current_time - detection_time > frames_cooldown):
        persons_detected += 1
        detection_time = current_time
    
    cv2.putText(frame, f"Count: {persons_detected}", (10, 45), cv2.FONT_HERSHEY_PLAIN, 3, (20, 255, 20), 2)

    cv2.imshow("frame", frame)
    cv2.imshow("avg", background)
    cv2.imshow("diff", diff)
    cv2.imshow("morph", morph)

    key = cv2.waitKey(PERIOD)

    if (key == ord('q')):
        break

print(f"People total: {persons_detected}")
cam.release()
cv2.destroyAllWindows()
cv2.waitKey(30)
```

## Вставляю себя на сцену

```python
import cv2
import mediapipe as mp
import numpy as np

# Initialize MediaPipe Selfie Segmentation
mp_selfie_segmentation = mp.solutions.selfie_segmentation
selfie_segmentation = mp_selfie_segmentation.SelfieSegmentation(model_selection=1)

# Loading and resizing image
image = cv2.imread('me.jpg')
height, width = image.shape[:2]
res = cv2.resize(image, (int(0.16 * width), int(0.16 * height)), interpolation=cv2.INTER_CUBIC)
image_rgb = cv2.cvtColor(res, cv2.COLOR_BGR2RGB)  # MediaPipe requires RGB input

# Process the image to get the segmentation mask
results = selfie_segmentation.process(image_rgb)
mask = results.segmentation_mask  

# Convert the mask to a binary image (0 or 255)
binary_mask = (mask > 0.5).astype(np.uint8) * 255

# Filter the mask
kernel = np.ones((5, 5), np.uint8)
binary_mask = cv2.morphologyEx(binary_mask, cv2.MORPH_CLOSE, kernel)
cv2.imwrite('mask.png', binary_mask)

background = cv2.imread('scene.jpg')
mask = cv2.imread('mask.png', cv2.IMREAD_GRAYSCALE)
center = (340, 180)
result = cv2.seamlessClone(res, background, mask, center, cv2.NORMAL_CLONE)

cv2.imwrite('result.jpg', result)

while(True):
    cv2.imshow('Seamless Cloning', result)
    key = cv2.waitKey(PERIOD)

    if (key == ord('q')):
        break

cv2.destroyAllWindows()
cv2.waitKey(30)

```
